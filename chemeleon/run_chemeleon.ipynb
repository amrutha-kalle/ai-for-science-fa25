```python
from chemeleon import Chemeleon
import gzip
import pickle
from pymatgen.core import Structure
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
```

```python
with gzip.open("../cifs_from_crystallm.pkl.gz", "rb") as f:
    cifs = pickle.load(f)
```

```python
chemeleon = Chemeleon.load_general_text_model()
```

```python
atoms_list_total = []
for cid, cif in cifs[20:]:
    structure = Structure.from_str(cif, fmt="cif")
    analyzer = SpacegroupAnalyzer(structure)
    conventional_structure = analyzer.get_conventional_standard_structure()
    n_atoms = len(conventional_structure)
    n_samples = 1
    output = chemeleon.sample(cif, n_atoms, n_samples)
    atoms_list_total.append((cif, cif, output))
    
```

```python
with gzip.open("chemeleon_output.pkl.gz", "wb") as f:
    pickle.dump(atoms_list_total, f)
```
